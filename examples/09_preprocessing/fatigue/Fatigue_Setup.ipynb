# Fatigue Preprocessing

This notebook will process metocean binned clusters into the format used by WEIS in the modeling options

```python
import pandas as pd
import numpy as np
import yaml, sys
try:
    import ruamel_yaml as ry
except Exception:
    try:
        import ruamel.yaml as ry
    except Exception:
        raise ImportError('No module named ruamel.yaml or ruamel_yaml')
```

```python
# Source: https://data.nrel.gov/submissions/241
df = pd.read_excel('HumboldtBay_Summary.xlsx',sheet_name=3,header=2)
```

```python
top10 = df.iloc[0:10]
top10
```

```python
top10.columns
```

```python
# Convert into names WEIS wants

weis_map = {
    'Cluster probability':'probability',
    'Wave Height (m)': 'wave_height_fatigue',
    'Wave Period (s)': 'wave_period_fatigue', 
    'Wind Dir. (deg)': 'wind_direction_fatigue',
    'Wave Dir. (deg)': 'wave_direction_fatigue',
    'Wind Speed (m/s)': 'wind_speed_fatigue',
}

top10.rename(columns = weis_map,inplace=True)

metocean_dict = {}

for val in weis_map.values():
    metocean_dict[val] = list(top10[val])


# list(top10['wind_speed_fatigue'])
    
    

```

```python
# yaml_string = yaml.dump(metocean_dict)
# print(yaml_string)


# f = open(fname, "w")
yaml=ry.YAML()
# yaml.default_flow_style = None
# yaml.width = float("inf")
# yaml.default_flow_style = True
yaml.indent(mapping=4, sequence=6, offset=3)
yaml.dump(metocean_dict,sys.stdout)
```
